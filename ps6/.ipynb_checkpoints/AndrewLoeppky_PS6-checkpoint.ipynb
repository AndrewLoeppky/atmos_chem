**24.3a)**

Prepare a plot of annual mean direct aerosol forcing $\Delta F (W/m^2)$, versus aerosol optical depth $\tau$ for different values of single-scatterinc albedo $\omega$ using equation (24.23).

$$
\Delta \bar{F} = -\frac{1}{2}F_0T_a^2(1-A_c)\omega\bar{\beta}\bar{\tau}\left[(1-R_s)^2 - \frac{2R_s}{\bar{\beta}}\left(\frac{1}{\omega}-1\right)\right]\tag{24.23}
$$

Consider $0 < \tau < 1.0$. Use values of $F_0$, $T_a$, $(1-A_c)$, $(1-R_s)$, and $\beta$ from Table 24.2. Consider $\omega$ values of 1.0, 0.9, 0.8, 0.7, and 0.5.  

<img src='table24-2.png' width='400'>

```python
############################################
######### CODE TO GENERATE PLOT ############
############################################

import numpy as np
import matplotlib.pyplot as plt

# Define variables
var_dict = {
    "F0": 1370,  # W/m2
    "one_minus_Ac": 0.4,
    "Ta": 0.76,
    "one_minus_R": 0.85,
    "R": 1 - one_minus_R,
    "beta": 0.29,
    "tau": np.linspace(0, 1, 100),
}

omega = (1, 0.9, 0.8, 0.7, 0.5)


def do_eq_2423(F0, one_minus_Ac, Ta, one_minus_R, R, beta, tau, omega):
    """solves Seinfeld 24.23 and returns change in radiative forcing"""
    deltaF = (
        -0.5
        * F0
        * Ta ** 2
        * one_minus_Ac
        * omega
        * beta
        * tau
        * (one_minus_R ** 2 - 2 * R / beta * (1 / omega - 1))
    )
    return deltaF


fig, ax = plt.subplots(1, 1, figsize=(10, 5))
for omega in omega:
    ax.plot(tau, do_eq_2423(**var_dict, omega=omega), label=f'$\omega = ${omega}' )
ax.legend()
ax.set_title('Radiative Forcing with Varying Optical Depth for Various Single Scattering Albedo Values')
ax.set_xlabel('Optical Depth')
ax.set_ylabel('$\Delta F$ (W/m^2)');
```

Aerosols with high albedo cool the surface by reflecting SW radiation back to space. Particles with lower albedo than the surface have a warming effect.
