### Seinfeld and Pandis 2.1
Methane (CH4) has an atmospheric mixing ratio of about 1800ppb. Carbon monoxide (CO) has a global average mixing ratio in the neighborhood of 100ppb. Global annual emissions of CH4 are about 600Tg(CH4)yr-1 (Table2.11); global annual CO sources are estimated as 2780Tg(CO)yr-1 (Table2.15). How can the atmospheric concentration of CH4 be almost 20 times higher than that of CO given their relative emissions? (Problem suggested by T.S.Dibble.)

We can solve this apparent paradox by examining equation 2.5 Sieinfeld and Prandis, the characteristic lifetime of a chemical species at steady state:

$$
\tau = \frac{Q}{P} = \frac{Q}{R} \tag{Seinfeld 2.5}
$$

Where $P = R$ ie the production and removal terms are equal for each species. Start by converting the given mixing ratios to mass for each species in the atmosphere (from the hint, there are $1.62 \times10^{20}$ moles of air in the atmosphere). The molar mass of methane is 16.04g/mol and the molar mass of carbon monoxide is 28.01g/mol.

$$
Q_{CH4} = \frac{1800 \space moles\space of\space CH_4}{10^9 moles\space of\space air }\times 1.62\times10^{20} moles\space of\space air \times 16.04g/mol = 4677\space Tg
$$

$$
Q_{CO2} = \frac{100 \space moles\space of\space CO}{10^9 moles\space of\space air }\times 1.62\times10^{20} moles\space of\space air \times 28.01g/mol = 453\space Tg
$$

Use these values to calculate $\tau$ for each species:

$$
\tau_{CH4} = \frac{Q_{CH4}}{P_{CH4}} = \frac{4677 Tg}{600 Tg/yr} \approx 8 years
$$

$$
\tau_{CO} = \frac{Q_{CO}}{P_{CO}} = \frac{453 Tg}{2780 Tg/yr} \approx 2 months
$$

$$
\frac{\tau_{CH4}}{\tau_{CO}} = 48
$$

While the emission rate for CO is 20 times higher, the characteristic lifetime is about 50 times shorter. So even though so much more CO is produced, it is removed from the atmosphere just as quickly so cannot accumulate to the same steady state concentration as methane. 

### Seinfeld and Prandis 2.3
The total estimated global sink of nitrous oxide (N2O) is 14.3TgNyr-1. The global mean N2O mixing ratio in 2011 was 324ppb. On the basis of these two values, estimate the mean lifetime of N2O in the atmosphere.

Use Seinfeld 2.5 and solve for $\tau$. First convert the mixing ratio to total mass of nitrogen (N2O) in the whole atmosphere:

$$
Q_N = 324ppb  \times \frac{1 N_2O}{10^9 air} \times 1.62 x10^{20}mol[air] \times 28.014 \frac{gN}{mol[N_2O]} = 1470\space Tg
$$

Solve 2.5:

$$
\tau = \frac{Q_N}{R} = \frac{1470 Tg}{14.3 Tg/yr} = \boxed{103 years}
$$

Based on this quick estimate, the mean lifetime of N2O is around 100 years, more or less consistent with the text. (Seinfeld quotes a study that found it to be 131 years)

### Jacob 8.1 \#2
Beryllium-7 is produced by cosmic rays in the stratosphere and upper troposphere. Similarly to 210Pb, it condenses immediately on preexisting aerosol particles. The global source of 7 Be is 150 g yr-1; 70% of that source is in the stratosphere and 30% is in the troposphere. Removal of 7 Be is by radioactive decay (half-life of 53 days) and by aerosol deposition (in the troposphere only). We assume that the troposphere and stratosphere are well-mixed reservoirs, that 7 Be is at steady state in each of these reservoirs, and that the transfer rate constant from the stratosphere to the troposphere is k ST = 0.8 yr -1. The total mass of 7 Be in the troposphere is estimated from observations to be about 3 g. Derive the residence time against deposition of 7 Be-carrying aerosols in the troposphere. You should find a value of 24 days.

Start with Seinfeld and Prandis 2.1, which describes the time-evolution of species $Q$ (in this case $^7$Be) and consider the stratosphere ($S$) and troposphere ($T$) as two separate systems each with their own source and sink terms:

$$
\frac{dQ}{dt} = F_{in} - F_{out} + P - R \tag{Seinfeld 2.1}
$$

#### Define each term for the troposphere and stratosphere:

We expect the exchange of $^7$Be at the tropopause to be diffusion-like (first order), where the transfer rate is dependent on the concentration difference with a proportionality constant $k_{ST}$ 

$
F_{in,T} = - F_{out,S} = k_{ST}(Q_S- Q_T)
$ 

#### Production (given)

$
P_{total} = 150\space g/yr
$

$
P_S = P_{total} \times 0.7
$

$
P_T = P_{total} \times 0.3
$

#### Removal
Removal follows exponential decay, calculate the time constant $a$ with time $t$ expressed in years

$$
Q_S(t) = Q_{S}(t_0)e^{-at} \tag{simple exponential decay}
$$

solve for $a$ given a half life of 53 days (0.145 years)
$$
Q_S(53) = \frac{1}{2}Q_S(0)
$$

$$
\frac{Q_S(0.145)}{Q_S(0)} = \frac{1}{2} = e^{-a(0.0145)}
$$

$$
a = -\frac{1}{0.145}ln\left(\frac{1}{2}\right) = 4.78\space years^{-1}
$$

take the time derivative to get the removal rate due to radioactive decay:

$$
\frac{d}{dt}(Q_Se^{-at}) = -Q_Sae^{-at}
$$

Since we assume steady state, we can set $t$ to any arbitrary value. Choose $t=0$ to take the exponential term to unity (also drop the negative to preserve the original sign convention in the stated problem). So the steady state removal term is:

$
R_T = Q_Sa
$

Removal in the troposphere has 2 modes, radioactive decay (which we just calculated) and deposition (which we need to calculate). Deposition is a first order process similar to transfer between the layers, but with it's own rate constant $k_D$. 

$
R_T = Q_Ta + k_DQ_T
$

Substitute into 2.1 to write a system of coupled DE's:

$$
\frac{dQ_S}{dt} = -k_{ST}(Q_S - Q_T) + 0.7P_{total} - Q_Sa \tag{1}
$$

$$
\frac{dQ_T}{dt} = k_{ST}(Q_S - Q_T) + 0.3P_{total} - (Q_Ta + k_DQ_T) \tag{2}
$$

List all given and derived variables thus far:

$
\frac{dQ_S}{dt} = 0 \space (steady \space state)
$

$
\frac{dQ_T}{dt} = 0 \space (steady \space state)
$

$
k_{ST} = 0.8 \space years^{-1}
$

$
Q_S = unknown
$

$
Q_T = 3 \space g
$

$
P_{total} = 150 \space g/year
$

$
a = 4.78 \space years^{-1}
$

$
k_D = unknown
$

Solve the system (2 equations 2 unknowns) via substitution:

**Equation 1**
$$
0 = -k_{ST}Q_S + k_{ST}Q_T + 0.7P_{total} - Q_Sa
$$

$$
(a + k_{ST})Q_S = k_{ST}Q_T + 0.7P_{total}
$$

$$
Q_S = \frac{k_{ST}Q_T + 0.7P_{total}}{a + k_{ST}}
$$

$$
\frac{(0.8)(3) + (0.7)(150)}{4.78 + 0.8} = 19.247 \space g
$$

**Equation 2**
$$
0 = k_{ST}(Q_S - Q_T) + 0.3P_{total} - Q_Ta - k_DQ_T 
$$

$$
k_D = \frac{k_{ST}(Q_S - Q_T) + 0.3P_{total} - Q_Ta}{Q_T}
$$

$$
\frac{(0.8)(19.247 - 3) + (0.3)(150) - (3)(4.78)}{3} = 14.55 \space years^{-1}
$$

Using Seinfeld 2.9, we can relate the transfer coefficients for each process to their respective residence times:

$$
\tau_1 = \frac{1}{k_1}, \tau_2 = \frac{1}{k_2} \tag{Seinfeld 2.9}
$$

So the residence time of $^7$Be with respect to deposition in the troposphere is:

$$
\tau_D = \frac{1}{k_D}
$$

$$
\frac{1}{14.55 \space years^{-1}} \approx \boxed{25\space days}
$$

Beryllium-7 deposited at the surface of the earth can be expected to have existed in the atmosphere for times on the order of about a month. Note the 1 day discrepency from Jacob is simply rounding error.
