### 15.1 
Rayleigh scattering is the irreducible minimum of light scattering owing to air molecules along an atmospheric sight path. The amound of scattered light varies as $\lambda^{-4}$. The Rayleigh scattering coefficient at three wavelengths is:

$$
b_{sg} = 25.9\left(\frac{293}{T}\right)p \quad \lambda = 450 \space nm
$$

$$
= 11.4\left(\frac{293}{T}\right)p \quad \lambda = 550 \space nm
$$

$$
= 5.8\left(\frac{293}{T}\right)p \quad \lambda = 650 \space nm
$$

where $T$ is in K and $p$ is in atmospheres. Calculate $b_{sg}$ at 550nm at sea level, 2km, and 3km altitude.

First convert sea level, 2km and 3km altitudes to pressure using the 1976 U.S Standard Atmosphere (table 1-5 Stull, Practical Meteorology, units converted)

$$
\begin{matrix}
  z \space (km) & T \space (K) & p \space (atm) \\
  0 & 288 & 1.00 \\
  2 & 275 & 0.78 \\
  3 & 269 & 0.69 
  \end{matrix}
$$

plug these numbers into the equation for $b_{sg}$ at 550nm

$$
b_{sg,0km} = 11.4\left(\frac{293}{288}\right) 1.00 = \boxed{1.16\cdot10^{-5} m^{-1}}
$$

$$
b_{sg,2km} = 11.4\left(\frac{293}{275}\right) 0.78 = \boxed{9.47\cdot10^{-6} m^{-1}}
$$

$$
b_{sg,3km} = 11.4\left(\frac{293}{269}\right) 0.69 = \boxed{8.57\cdot10^{-6} m^{-1}}
$$

## 15.2
Light extinction can be divided into the sum of its scattering and absorbtion components as follows

$$
b_{ext} = b_{sg} + b_{ag} + b_{sp} + b_{ap}
$$

where $b_{ext}$ = light extinction coefficient, $b_{sg}$ = Rayleigh scattering (light scattering by molecules of air), $b_{ag}$ = light absorbtion because of gasses (mainly NO$_2$), $b_{sp}$ = light scattering by particles, $b_{ap}$ = light absorbtion by particles. The scattering and absorbtion by gases ($b_{sg}$ and $b_{ag}$) can be calculated knowing the air pressure (altitude) and temperature, and the concentration of NO$_2$, respectively. To deal with the particle-related extinction ($b_{sp}$ and $b_{ap}$), a standard approach is to allocate portions of the extinction to each species of the mixture and then summarize the contributions to arrive at the total particle-related extinction. With this approach, one sums up the individual contributions

$$
b_{sp} = e_{sulfate}[sulfate] + e_{nitrate}[nitrate] + e_{OC}[organic \space carbon] + e_{soil}[soil] + e_{coarse}[coarse]
$$

and, for absorption of light by particles

$$
b_{ap} = e_{BC}[black \space carbon]
$$

where the $e$ values are the extinction efficiencies. The units of $e$ are M m$^{-1}$ per $\mu m^{-3}$; hence m$^2g^{-1}$. Extinction efficiencies depend on the size distribution and the molecular composition of the sulfates, nitrates, and organic carbon. The extinction efficiencies for hygroscopic substances (sulfate, nitrate, and organic carbon) are dependent on the relative humidity. Values are usually reported for dry particles; the uptake of water can multiply the given sulfate and nitrate efficiencies manyfold at high relative humidities. The effect of water uptake on the organic carbon efficiency is not as well established as that for the inorganic salts. Ranges of dry extinction efficiencies are

$$
e_{sulfate} = 1.5 - 4 m^2g^{-1}
$$

$$
e_{nitrate} = 2.5 - 3 m^2g^{-1}
$$

$$
e_{OC} = 1.8 - 4.7 m^2g^{-1}
$$

$$
e_{soil} = 1 - 1.25 m^2g^{-1}
$$

$$
e_{coarse} = 0.3 - 0.6 m^2g^{-1}
$$

$$
e_{BC} = 8 - 12 m^2g^{-1}
$$

Calculate the visual range of an atmosphere at a 0.02 contrast ratio for which $e_{sulfate}$ = 3m$^2$g$^{-1}$, $e_{nitrate}$ = 3m$^2$g$^{-1}$, $e_{OC}$ = 4m$^2$g$^{-1}$, and $e_{BC}$ = 10m$^2$g$^{-1}$, and for which

$$
[Sulfate] = 20 \mu g m^{-3}
$$

$$
[Nitrate] = 5 \mu g m^{-3}
$$

$$
[Organic \space Carbon] = 25 \mu g m^{-3}
$$

$$
[Black \space Carbon] = 7.5 \mu g m^{-3}
$$

Approach each component of $b_{ext}$ one by one, starting with Rayleigh scattering, assuming the same $\lambda = 550\space nm$, $T=288\space K$ and $p=1 \space atm$ as the previous question

$$
b_{sg} = 11.4\left(\frac{293}{288}\right) 1.00 = 1.16\cdot10^{-5} m^{-1}
$$

Calulate $b_{sp}$ using the equation given above

$$
b_{sp} = e_{sulfate}[sulfate] + e_{nitrate}[nitrate] + e_{OC}[organic \space carbon]
$$

$$
b_{sp} = (3m^2g^{-1})(20 \mu g m^{-3}) + (3m^2g^{-1})(5 \mu g m^{-3}) + (4m^2g^{-1})(25 \mu g m^{-3}) + 
$$

$$
b_{sp} = 1.75\cdot 10^{-4} m^{-1}
$$

for light absorption by black carbon:

$$
b_{ap} = e_{BC}[black \space carbon]
$$

$$
b_{ap} = (10m^2g^{-1})(7.5 \mu g m^{-3})
$$

$$
b_{ap} = 7.5\cdot 10^{-5} m^{-1}
$$

assuming negligible contribution from NO$_2$, add the components together to get $b_{ext}$

$$
b_{ext} = 1.16\cdot10^{-5} m^{-1} + 1.75\cdot 10^{-4} m^{-1} + 7.5\cdot 10^{-5} m^{-1}
$$

$$
b_{ext} = 2.616\cdot 10^{-4}m^{-1}
$$

Finally, for a contrast ratio $C_V = 0.02$, we can use Seinfeld 15.36 to calculate the visual range at the given aerosol loading

$$
x_v = \frac{3.912}{b_{ext}}\tag{Seinfeld 15.36}
$$

$$
x_v = \frac{3.912}{2.616\cdot 10^{-4}m^{-1}} = 14954m
$$

$$
\boxed{x_v = 15 \space km}
$$
